# Contours detections

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_img = cv2.imread('download.png')
car_img = cv2.cvtColor(car_img,cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(car_img) 
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    car_image_copy = img.copy()
    
    noplate_rect = cascade.detectMultiScale(car_image_copy,scaleFactor = 1.3,
                                           minNeighbors=2)
    
    for (x,y,w,h) in noplate_rect:
        cv2.rectangle(car_image_copy,(x,y),(x+w,y+h),(255,0,0),3)
    return car_image_copy
```

```python
result = detect_plate(car_img)
```

```python
display(result)
```
