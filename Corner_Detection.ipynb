# CORNER DETECTION
    # Corner is a point whose local neighborhood stands in   two dominant and diffrent       edge directions.
    #There are diffrent corner detection algorithm  , popular algorithms are : 
        1- Harris Corner detection.
        2- Shi - Tomasi Corner Detection. 
    

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('flat_cheese.jpeg') 
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_cheess = cv2.imread('real_chess_board.jpg')
real_cheess = cv2.cvtColor(real_cheess,cv2.COLOR_BGR2RGB)
plt.imshow(real_cheess)
```

```python
gray_real_chess = cv2.cvtColor(real_cheess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray_flat_chess
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dest = cv2.cornerHarris(src =gray,blockSize=2,ksize = 3,k=0.04)
```

```python
dest = cv2.dilate(dest,None) 
```

```python
flat_chess[dest>0.01*dest.max()] = [255,0,0] #RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess) 
dest = cv2.cornerHarris(src=gray,blockSize=2,ksize = 3,k=0.04)
```

```python
dest = cv2.dilate(dest,None)
```

```python
real_cheess[dest>0.01*dest.max()] = [255,0,0]
```

```python
plt.imshow(real_cheess)
```

```python
#PART2 Shi - Tomasi Corner Detection. 
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,5,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners :
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,20,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_cheess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(real_cheess)
```
