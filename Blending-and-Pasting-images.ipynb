```python
img1 = cv2.imread('opencvlogo.png')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('img1.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1 = cv2.resize(img1,(600,600))
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
x_offset = 1200 - 600
y_offset = 700  - 600
```

```python
rows,cols,channels = img1.shape
```

```python
roi = img2[y_offset:700,x_offset:1200]
```

```python
plt.imshow(roi)
```

```python
#CREATING A MASK 
```

```python
img2gray = cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
plt.imshow(mask_inv)
```

```python
mask_inv.shape
```

```python
white_background = np.full(img1.shape,255,dtype=np.uint8)
```

```python
white_background.shape
```

```python
plt.imshow(white_background)
```

```python
#white_background
```

```python
bk = cv2.bitwise_or(white_background,white_background,mask=mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img1,img1,mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi,fg)
```

```python
plt.imshow(final_roi)
```

```python
small_img = final_roi
large_img = img2
```

```python
large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]] = small_img
```

```python
plt.imshow(small_img)
```
