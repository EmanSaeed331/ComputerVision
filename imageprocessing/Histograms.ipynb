```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse   = cv2.imread('horse.jpeg') #ORIGINAL BGR OPENCV
show_horse   = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) #CONVERTED TO RGB FOR SHOW

reinbow      = cv2.imread('reinbow.jpeg')
show_reinbow = cv2.cvtColor(reinbow,cv2.COLOR_BGR2RGB)

bricks       = cv2.imread('bricks.jpeg')
show_bricks = cv2.cvtColor(bricks,cv2.COLOR_BGR2RGB) 
```

```python
plt.imshow(show_horse)
plt.imshow(show_reinbow)
plt.imshow(show_bricks)
```

```python
gorilla = cv2.imread('gorilla.jpeg',0)
```

```python
def display_img(img , cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display_img(gorilla,cmap='gray')
```

```python
gorilla.shape
```

```python
 #CREATE A HISTOGRAM AND EQUILIZED HISTOGRAM AND CALCULATE THE DIFFRENCE.
```

```python
hist_values = cv2.calcHist([gorilla],channels=[0],mask=None,histSize = [256],ranges =[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
display(eq_gorilla)
```

```python
display_img(eq_gorilla,cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_gorilla],channels=[0],mask=None,histSize = [256],ranges =[0,256])
```

```python
plt.plot(hist_values)
```
